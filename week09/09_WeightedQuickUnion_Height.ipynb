```
import graphviz
def drawUF(a):
    dot = graphviz.Digraph(comment='Nada')
    n = len(a)
    for i in range(n):
        dot.node(str(i), str(i))
    for i in range(n):
        if a[i] > 0:
            dot.edge(str(i), str(a[i]))
    dot.graph_attr['rankdir'] = 'BT'
    return dot
```

# Union Find

## Quick Union Ponderado

```
def find(id, a):
    while id[a] > 0:
        a = id[a]

    return a

def union(id, a, b):
    parentA = find(id, a)
    parentB = find(id, b)
    if parentA == parentB:
        return
    if -id[parentA] < -id[parentB]:
        id[parentA] = parentB
    elif -id[parentA] > -id[parentB]:
        id[parentB] = parentA
    else:
        id[parentA] -= 1
        id[parentB] = parentA
```

```
id = [-1 for _ in range(10)]; print(id); drawUF(id)
```
