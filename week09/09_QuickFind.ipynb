```
#Para dibujar
import graphviz
def drawUF(a):
    dot = graphviz.Digraph(comment='Nada')
    n = len(a)
    for i in range(n):
        dot.node(str(i), str(i))
    for i in range(n):
        if i != a[i]:
            dot.edge(str(i), str(a[i]))
    dot.graph_attr['rankdir'] = 'BT'
    return dot
  
```

**QUICK FIND**

```
def quickFind(id, a):
    return id[a]

def quickFindUnion(id, a, b):
    parentA = quickFind(id, a)
    parentB = quickFind(id, b)
    if parentA == parentB:
        return
    for i in range(len(id)):
        if id[i] == parentA:
            id[i] = parentB
            
id = [i for i in range(10)]

print(id)
quickFindUnion(id, 3, 4);print(id)
quickFindUnion(id, 4, 9);print(id)
quickFindUnion(id, 8, 0);print(id)
quickFindUnion(id, 2, 3);print(id)
quickFindUnion(id, 5, 6);print(id)
quickFindUnion(id, 5, 9);print(id)
quickFindUnion(id, 7, 3);print(id)
quickFindUnion(id, 4, 8);print(id)
quickFindUnion(id, 6, 1);print(id)
```
