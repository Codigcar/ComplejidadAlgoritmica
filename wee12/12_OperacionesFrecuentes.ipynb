```
def loadAdjMatGraph(filename):
    G = []
    with open(filename) as file:
        for line in file:
            G.append([int(num) for num in line.split(' ')])

    return G
```

```
import graphviz as gv

def drawAdjMatGraph(G):
    dot = gv.Graph(strict=True)
    n = len(G)
    for u in range(n):
        for v in range(n):
            w = G[u][v]
            if w != 0:
                dot.edge(str(u), str(v), label=str(w))
    
    dot.graph_attr['rankdir'] = 'BT'
    return dot
```

```
G = loadAdjMatGraph('grafo.am')

drawAdjMatGraph(G)
```

# Recorrer vertices de un grafo en matriz de adyacencia

```
n = len(G) # necesitamos saber cuantos vértices hay

for u in range(n):
    print("Vértice", u)
```

# Recorrer todos los arcos de un grafo en matriz de adyacencia

```
for u in range(n):
    for v in range(n):
        w = G[u][v]
        if w != 0:
            # tenemos el arco de u a v con peso w.
            print("Arco de", u, "a", v, "con peso", w)
```

# Recorrer todos los arcos de salida de un vértice particular en una matriz de adyacencia

```
# sea el vértice u = 5
u = 5
for v in range(n):
    w = G[u][v]
    if w != 0:
        print("Arco de", u, "a", v, "con peso", w)
```

# Hagamos lo mismo para Lista de Adyacencia

```
%%writefile grafo.al
2,8 3,8
5,8 6,3
0,8 5,1
0,8
7,8
1,8 2,1 6,2
1,3 5,2 7,6
4,8 6,6
```

```
def loadAdjListGraph(filename):
    G = []
    with open(filename) as file:
        for line in file:
            #print(line)
            temp = line.strip().split(' ')
            #print(temp)
            temp = [p.split(',') for p in temp]
            #print(temp)
            temp = [(int(x), int(y)) for x, y in temp]
            #print(temp)
            G.append(temp)

    return G
        
```

```
G = loadAdjListGraph('grafo.al')
print(G)
```

```
def drawAdjListGraph(G):
    dot = gv.Graph(strict=True)
    n = len(G)
    for u in range(n):
        for v, w in G[u]:
            dot.edge(str(u), str(v), label=str(w))
    
    dot.graph_attr['rankdir'] = 'BT'
    return dot
```

```
drawAdjListGraph(G)
```

# Recorrer vértices de un grafo en lista de adyacencia

```
n = len(G)
for u in range(n):
    print("vértice", u)
```

# Recorrer todos los arcos de un grafo en lista de adyacencia

```
for u in range(n):
    for v, w in G[u]:
        print("arco de", u, "a", v, "con peso", w)
```

# Recorrer arcos de salida de un vértice particular

```
# sea el vértice u = 5
u = 5
for v, w in G[u]:
    print("arco de", u, "a", v, "con peso", w)
```
